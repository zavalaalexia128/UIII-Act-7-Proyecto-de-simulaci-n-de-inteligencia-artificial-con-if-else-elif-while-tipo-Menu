## Clase Relojero

```python
def menu():
    print("\n--- MENU ---")
    print("1. Reloj no funciona")
    print("2. Cambiar bateria")
    print("3. Revisar cristal")
    print("4. Salir")
    opcion = int(input("Ingrese una opcion: "))
    return opcion

opcion = 0

while opcion != 4:

    opcion = menu()

    if opcion == 1:

        tipo = input("¿El reloj es digital o mecanico?: ").lower()

        if tipo == "digital":
            print("IA: Revisar bateria.")
        elif tipo == "mecanico":
            print("IA: Revisar engranes.")
        else:
            print("Tipo no valido.")

    elif opcion == 2:

        marca = input("Ingrese la marca del reloj: ")

        if marca == "Casio":
            print("IA: Usa bateria CR2016.")
        else:
            print("IA: Revisar manual.")

    elif opcion == 3:

        cristal = input("¿El cristal esta roto? (si/no): ").lower()

        if cristal == "si":
            print("IA: Cambiar cristal.")
        else:
            print("IA: Cristal en buen estado.")

    elif opcion == 4:
        print("Programa finalizado.")
        print("Programa Realizado por Zavala Medina Alexia Paola NL= 59")

    else:
        print("Opcion invalida.")
```
