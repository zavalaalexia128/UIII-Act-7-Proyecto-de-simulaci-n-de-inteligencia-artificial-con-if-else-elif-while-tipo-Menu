## Clase Relojero

```python
print("SISTEMA DE ENTRENAMIENTO VIRTUAL 'RELOJERO-INTELLIGENCE v2.0'")
print("Bienvenido, Cliente. El asistente relojero esta listo para ayudarte.")

sesion_activa = "si"

# El ciclo while mantiene la IA activa hasta finalizar el servicio
while sesion_activa == "si":

    print("\n*********************************")
    print("MENU DEL CLIENTE:")
    print("1. Solicitar herramienta para reparacion")
    print("2. Reportar una falla para diagnostico")
    print("3. Finalizar servicio")
    print("*********************************")

    comando = input("¿Que necesitas hacer? (1/2/3): ")

    # Estructura if-elif-else para procesar el menu
    if comando == "1":

        print("\n--- ASISTENTE DE HERRAMIENTAS ---")

        tipo_trabajo = input("¿El trabajo es 'ajustar', 'limpiar' o 'reparar'?: ").lower()

        # La IA decide la herramienta usando logica condicional
        if tipo_trabajo == "ajustar":
            print("Asistente IA: Necesitaras el destornillador de precision.")

        elif tipo_trabajo == "limpiar":
            print("Asistente IA: Usa el cepillo suave y el liquido limpiador.")

        elif tipo_trabajo == "reparar":
            print("Asistente IA: Trae las pinzas y revisa los engranes.")

        else:
            print("Asistente IA: Esa tarea no esta registrada en el sistema relojero.")

    elif comando == "2":

        print("\n--- DIAGNOSTICO DE FALLAS ---")

        gravedad = input("¿Cual es el nivel de daño del reloj? (bajo/medio/alto): ").lower()

        # Simulacion de razonamiento condicional
        if gravedad == "alto":
            print("Asistente IA: ¡CRITICO! El mecanismo interno debe repararse de inmediato.")

        elif gravedad == "medio":
            print("Asistente IA: Ajusta las piezas y revisa la bateria.")

        else:
            print("Asistente IA: Solo necesita mantenimiento basico.")

    elif comando == "3":

        print("\nAsistente IA: Guardando informacion del servicio... ¡Gracias por su visita!")

        # Cambiamos la variable para terminar el ciclo while
        sesion_activa = "no"

    else:

        # Manejo de errores del menu
        print("\n[Error del sistema]: Comando no reconocido. Intenta de nuevo.")

print("\n--- PROGRAMA FINALIZADO ---")
print("Programa realizado por Zavala Medina Alexia Paola")
```
